```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
logs = np.load('../cnn-keras-update/results/2017-01-21_23:20:48/logs.npy')
```

```python
def plot_logs(hist, metric='loss', title='', figsize=(20,5)):
    plt.figure(figsize=figsize)
    x = list(range(len(hist)))
    if metric == 'loss':
        plt.plot(x, hist[:,0], label='train')
        plt.plot(x, hist[:,5], label='val')
    elif metric == 'loss_age':
        plt.plot(x, hist[:,1], label='train')
        plt.plot(x, hist[:,6], label='val')
    elif metric == 'loss_gender':
        plt.plot(x, hist[:,2], label='train')
        plt.plot(x, hist[:,7], label='val')
    elif metric == 'acc_age':
        plt.plot(x, hist[:,3], label='train')
        plt.plot(x, hist[:,8], label='val')
    elif metric == 'acc_gender':
        plt.plot(x, hist[:,4], label='train')
        plt.plot(x, hist[:,9], label='val')
    plt.xlabel('epoch')
    plt.ylabel(metric)
    plt.ylabel(title)
    plt.legend()
```

```python
plot_logs(logs, metric='acc_age', title='accuracy [age]', figsize=(15,3))
plot_logs(logs, metric='acc_gender', title='accuracy [gender]', figsize=(15,3))
```

```python
plot_logs(logs, metric='loss', title='loss [combined]', figsize=(15,6))
plot_logs(logs, metric='loss_age', title='loss [age]', figsize=(15,3))
plot_logs(logs, metric='loss_gender', title='loss [gender]', figsize=(15,3))
```
